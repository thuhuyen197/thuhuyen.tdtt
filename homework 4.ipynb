W4A18 Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a,b= map(int,input().split())
for i in range(1,min(a,b)+1):
    if a%i==0 and b%i==0:
        print(i,end =" ")
```

W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n= int(input())
while n>1:
    if n%2!=0:
        print('No')
        break
    n //=2
else:
    print('Yes')
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n= int(input())
chan= 0
le = 0
while n!=0:
    dem=n%10
    if dem%2==0:
        chan +=1
    else:
        le +=1
    n//=10
print(chan,le)
```

Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 

```python
n=int(input())
s=0
k=0
while s+(k+1)<n:
    k+=1
    s+=k
print(k)

```

W4A24
Nhập vào A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 



```python
A=int(input())
n=0
s=0
while s<=A:
    n+=1
    s+=1/n
print(n)
```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 


```python
numbers=[]
while True:
    n=int(input())
    if n==-1:
        break
    numbers.append(n)
if len(numbers) ==0:
    print("không có phần tử nào")
else:
    print(max(numbers))
    print(min(numbers))
```

W4A27 [Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python
A=int(input())
a,b=1,1
while b<=A:
    a,b=b,a+b
print(a)

```

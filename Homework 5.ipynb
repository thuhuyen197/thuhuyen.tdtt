W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.


```python
def doi_xung(n):
    so = n
    n_rev = 0
    
    while n > 0:
        x = n % 10
        n_rev = n_rev * 10 + x
        n //= 10
    
    return so == n_rev

n = int(input())
print(doi_xung(n))
```

W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên

```python

def ki_dieu(n):
    while n >= 10:
        s = 0
        while n > 0:
            s += n % 10
            n //= 10
        n = s
    return n == 1

n = int(input())
print(ki_dieu(n))
```

W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def dem_0(lst):
    count = 0
    for n in lst:
        while n > 0:
            if n % 10 != 0:
                count += 1
            n //= 10
    return count

lst= list(map(int,input().split()))
print (dem_0(lst))
```

W5A30 Viết hàm tính hệ số nhị thức C(n, k)

```python
#C(n,k)=[n(n−1)(n−2)...(n−k+1)​]/k
def C(n,k):
    if k>n:
        return 0
    x =1
    for i in range (1,k+1):
        x = x* (n-i+1)//i
    return x
n = int(input())
k = int(input())
print(C(n,k))
```

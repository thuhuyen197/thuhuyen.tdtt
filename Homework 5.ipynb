W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.


```python
def doi_xung(n):
    so = n
    n_rev = 0
    
    while n > 0:
        x = n % 10
        n_rev = n_rev * 10 + x
        n //= 10
    
    return so == n_rev
```

```python

def ki_dieu(n):
    while n >= 10:
        s = 0
        while n > 0:
            s += n % 10
            n //= 10
        n = s
    return n == 1
```

W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def dem_0(list):
    count = 0
    for n in list:
        while n > 0:
            if n % 10 != 0:
                count += 1
            n //= 10
    return count
```

W5A30 Viết hàm tính hệ số nhị thức C(n, k)

```python
#C(n,k)=[n(n−1)(n−2)...(n−k+1)​]/k
def C(n,k):
    if k>n:
        return 0
    x =1
    for i in range (1,k+1):
        x = x* (n-i+1)//i
    return x
```

W5A31 Viết hàm tính tổng bình phương các số từ 1 đến n

```python
def tong_binhphuong(n):
    sum = 0
    for i in range (1,n+1):
        sum += i*i
    return sum
```

W6A1:Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]

```python
l = map(int,input().split())
rs = []
for i in l:
    if i not in rs:
        rs.append(i)
print(rs)
```

Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu
tới vị trí i), giữ nguyên thứ tự

```python
l = map(int,input().split())
s = 0
for i in l:
    s +=i
    rs.append(s)
print(rs)
```

W6A4 :Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key
lặp lại, thêm vào list của key đó

```python
pairs = input().split()
rs = {}
for p in pairs:
    k,v = p.split(':')
    if k not in rs:
        rs[k] = []
    rs[k].append(v)
rs
```

W6A5
Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 
'zeros': ...} (đếm số >0, <0, =0)


```python
l = map(int,input().split())
d = {'positives':0,'negatives':0,'zeros':0}
for x in l:
    if x>0:
        d['positives']+=1
    elif x<0:
        d['negatives']+=1
    elif x==0:
        d['zeros']+=1

print(d)
```
